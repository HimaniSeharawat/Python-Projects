```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_excel(r'C:\Users\hp\Desktop\Dr Sam Min\Assignments\31 August 2021\curriculum.xlsx')
```

```python
df.Description = np.where(df.Description.isnull()," ",df.Description)
df=df.groupby(['Department'])['Description'].apply(','.join).reset_index()
df['Description']
```

```python
df.to_excel (r'C:\Users\hp\Desktop\Dr Sam Min\Step02.xlsx', index = False, header=True)
```

```python
tfidf_vectorizer = TfidfVectorizer()
sparse_matrix = tfidf_vectorizer.fit_transform(df['Description'])
doc_term_matrix = sparse_matrix.toarray()
tgt_transform = tfidf_vectorizer.transform(df['Description']).toarray()
tgt_cosine = cosine_similarity(doc_term_matrix)
similaritymatrixofdepartment = pd.DataFrame(tgt_cosine,columns=df['Department'],index=df['Department'])

print(similaritymatrixofdepartment)
```

```python
similaritymatrixofdepartment.to_excel (r'C:\Users\hp\Desktop\Dr Sam Min\Step04.xlsx', index = False, header=True)
```
