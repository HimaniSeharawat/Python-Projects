```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid_deaths_usafacts.csv')
df
```

```python
df = df.loc[df['State'] == 'IA']
df
```

```python
df1 = df.melt(id_vars=[ "countyFIPS","County Name",'State',"StateFIPS"], 
        var_name="Date")
```

```python
df1
```

```python
df1.drop(columns=['countyFIPS','State','StateFIPS'],inplace=True)
df1
```

```python
df2 = df1.sort_values(by = ['County Name', 'value'], ascending = [True, True],na_position = 'first')
df2.reset_index(drop=True, inplace=True)
```

```python
df2.rename(columns={'County Name':'County','value': 'cum_case'}, inplace=True)
```

```python
df2.to_csv('Output.csv', index=False)
```
