```python

# Make small testing username list 
test_user = usernames[:180]  # ['0AshleyMay0', '0Lori0','0Lori0', '0hkaye', '1125dmw', '123Carol', '13luckythirteen', '14bellesfleurs', '1Acarbine',
#      '1HawtMama71', '1Silentangel', '1StarGazer1', '1lali', '1prettyscrapper', '20TayTay17', '212mom', '25tessa', '26Minnie', 
#     '2beAquarius', '2lovebeauty','305MIA',36524,'3kmmoore','3tboys3','40thpony','45plus','4EverJazz','4LoveorMoney',
#     '4rzoo','4sistersalison','55goingon45','6come6','6ixGirl','6jordans','740p','760SweeTt','77tonyatijLa','7waystowin',
#     '8319Princess','96chocolate96','96chocolate96',97068463,'98Kimmy','A1228','A147','A1Pia','ACEMILL','ADKer','ADRl',
#     'BBalexandra', 'BBrooklyn', 'BENNETT40', 'BGChris', 'BGlossy', 'BHarts', 'BLovedDestroya', 'BOSSBMUA', 'BPATEL1', 'BR25',
#     'BReady716', 'BabiKhae', 'Babs2014', 'Babserella', 'BabyFish', 'BabyJekyll', 'Babyaj', 'Babypeachist', 'Backchloe1', 
#     'BadGalBritty', 'Badgalcincin', 'Bail3yB3auty', 'Baisant', 'BajanIslandGirl', 'Bakhtu', 'BalmSquad22', 'Balrog','CANNAV',
#     'CAT3MOM', 'CAcay', 'CC3616', 'CCHL', 'CCHelpmeWMakeup', 'CConlan', 'CDLL', 'CEOHaley', 'CElez', 'CJM2017', 'CKohl',
#     'CKris', 'CLSMengotti', 'CLidi','D0llface8455', 'DABOSSY1', 'DAGC', 'DEEBOMB', 'DEL28', 'DFine', 'DLCN', 'DMatijas', 
#     'DWine', 'DZuniga', 'Dadoune', 'Daenique', 'Dafne', 'DaintyDiva', 'Daise66', 'Daishashanice', 'Daisy108', 'DaisyAmy',
#     'DaisyMaryjane', 'DaisySG', 'Daisydreamer', 'Dalalous', 'DalilaI', 'Dallasgirl88', 'DanaBritt', 'DanaFritchie', 
#     'DanaP73', 'DancininDaizys', 'DaneeBT', 'DaniBlu08', 'DaniEastLA', 'Dania0715', 'Daniahamdan1', 'DanicaS', 'Daniellabella23',
#     'EBond', 'EChristine29', 'EELINAA', 'ELNene17', 'EMLynley', 'EMLynley', 'EPaden', 'ESAMAKEUP', 'EdenLake', 'EedEed', 'Ejh77', 
#      'El67','FNaveed', 'Fabbzi', 'Fabhuntress', 'FaceTimeFee', 'Facebeatfashion', 'FairTexan', 'FairyBrooke', 'Faith2bm', 
#      'Fallingsnow', 'FalsiesFan', 'Fanny24', 'Fantasme']


# Get the user on the basis of username
def get_url(u):
    """Generate Url """
    temp = "https://www.sephora.com/users/{}"
    url = temp.format(u)
    return url

# Extrat the data using soup
def getAllData(soup):
    
    # username of the shephora user
    name = ''
    temp_name = soup.find(class_='css-9pkcf6 eanm77i0') # tag css-9pkcf6 eanm77i0
    if temp_name is not None: 
        name = temp_name.find('b').text
    else:
        name = ""
    
    # Profile url of user
    profile_url = ''
    url = soup.find('div',{'class':'css-1n8s0sw eanm77i0'})
    if url is not None:
        profile_url = cssutils.parseStyle(url['style'])['background-image'].replace('url(', '').replace(')', '')
    else:
        profile_url = ""

    # Inider 
    insider = "" 
    ins = soup.find('div',{'class':'css-s83hmn eanm77i0'})
    if ins is not None:
        insider = ins.find('img')['alt']
    else:
        insider = ""
    
    # Types of Rokiees
    rookie = ""
    rook = soup.find('div',{'class':'css-s83hmn eanm77i0'})
    if rook is not None:
        rookie = rook.find_all('img')[-1]['src'].split('/')[-1][5:-4].upper()
    else:
        rookie = ""
    
    # Is user have default profile or not
    is_default = ""
    if profile_url is not None:
        is_default =  'Yes' if 'default-avatar' in profile_url else "No"
    else:
        is_default = ""

    sephoraRecord = (name, profile_url, insider, rookie,is_default)
    return sephoraRecord

# main function to run the all the above function to get the desired output

def main(test_user):
    sephoraRecord = []
    for ii in test_user:
        url = get_url(str(ii))
        
        
# Extract required User  data
        driver = webdriver.Chrome(executable_path= r"D:\\Techeron\\ChromeDriver\\chromedriver.exe")
        driver.get(url)
        driver.set_page_load_timeout(3600)
        time.sleep(random.randint(5,12))
        html = driver.page_source  
        soup = BeautifulSoup(html, "html.parser")
        record = getAllData(soup)
        sephoraRecord.append(record) 
  
      
  #  Save the Data into CSV file
    csv_file_name = 'UserDetails.csv'
    with open(csv_file_name, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(["username","profile_url","Insider","Rookie","Is_default"])
        writer.writerows(sephoraRecord)
```

```python
main(test_user)
```
